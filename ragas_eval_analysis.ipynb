0. Importing libraries and dependencies

```python
import pandas as pd
import json
```

1. Importing RAGAS evaluation report for Recursive chunking methodology

```python
df_rec = pd.read_csv("run_logs/eval_2026-07-21T18-40-31+00-00_recursive_ragas.csv")
print(f"RAGAS evaluation - Recursive Chunking Strategy")

# Overall averages per metric
print(df_rec[["faithfulness", "context_precision", "answer_relevancy"]].mean())

# Worst-scoring questions per metric
print(df_rec.sort_values("faithfulness").head(5)[["user_input", "faithfulness"]])
```

2. Importing RAGAS evaluation report for Fixed chunking methodology

```python
df_fix = pd.read_csv("run_logs/eval_2026-07-22T11-16-26+00-00_fixed_ragas.csv")
print(f"RAGAS evaluation - Fixed Chunking Strategy")

# Overall averages per metric
print(df_fix[["faithfulness", "context_precision", "answer_relevancy"]].mean())

# Worst-scoring questions per metric
print(df_fix.sort_values("faithfulness").head(5)[["user_input", "faithfulness"]])
```

3. Side-by-Side comparison

```python
metrics = ["faithfulness", "context_precision", "answer_relevancy"]
ragas_comparison = pd.DataFrame({
    "fixed": df_fix[metrics].mean(),
    "recursive": df_rec[metrics].mean(),
})
# IR metrics from the eval_*.json logs
with open("run_logs/eval_2026-07-22T11-16-26+00-00_fixed.json") as f:
    ir_fixed = json.load(f)["ir_metrics"]["overall"]
with open("run_logs/eval_2026-07-21T18-40-31+00-00_recursive.json") as f:
    ir_recursive = json.load(f)["ir_metrics"]["overall"]

ir_comparison = pd.DataFrame({
    "fixed": ir_fixed,
    "recursive": ir_recursive,
}).drop("n")  # drop item count, not a score

# combine both into one table
comparison = pd.concat([ir_comparison, ragas_comparison])

print(comparison)
```
